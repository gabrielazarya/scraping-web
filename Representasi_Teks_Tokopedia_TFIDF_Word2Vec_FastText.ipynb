# Representasi Teks Ulasan Tokopedia
Metode yang digunakan:
1. TF-IDF
2. Word2Vec
3. FastText

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from gensim.models import Word2Vec, FastText
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Load data ulasan
df = pd.read_csv("Ulasan_Lengkap_Tokopedia.csv")
ulasan = df["Ulasan"].dropna().astype(str).tolist()
ulasan[:500] 
```

## 1. TF-IDF Vectorization

```python
vectorizer = TfidfVectorizer(max_features=100)
tfidf_matrix = vectorizer.fit_transform(ulasan)

# Konversi ke DataFrame
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=vectorizer.get_feature_names_out())
tfidf_df.head()
```

## 2. Word2Vec Embedding

```python
# Tokenisasi sederhana
tokenized_ulasan = [ul.split() for ul in ulasan]

# Train model Word2Vec
w2v_model = Word2Vec(sentences=tokenized_ulasan, vector_size=100, window=5, min_count=1, workers=4)
w2v_vector = [sum([w2v_model.wv[word] for word in words if word in w2v_model.wv], 0) / len(words) for words in tokenized_ulasan]

# Konversi ke DataFrame
w2v_df = pd.DataFrame(w2v_vector)
w2v_df.head()
```

## 3. FastText Embedding

```python
# Train model FastText
ft_model = FastText(sentences=tokenized_ulasan, vector_size=100, window=5, min_count=1, workers=4)
ft_vector = [sum([ft_model.wv[word] for word in words if word in ft_model.wv], 0) / len(words) for words in tokenized_ulasan]

# Konversi ke DataFrame
ft_df = pd.DataFrame(ft_vector)
ft_df.head()
```
